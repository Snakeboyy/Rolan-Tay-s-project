```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

file_path = "./red_wine.csv"
red_wine_data = pd.read_csv(file_path)

```

```python
print("First 10 Rows:")
print(red_wine_data.head(10))
```

```python
print("Last 20 Rows:")
print(red_wine_data.tail(20))
```

```python
print(red_wine_data.info())
```

```python
print(red_wine_data.describe())
```

```python
plt.figure(figsize=(10, 10))
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 
            'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 
            'pH', 'sulphates', 'alcohol', 'quality']

for i, feature in enumerate(features):
    plt.subplot(4, 3, i+1)
    plt.hist(red_wine_data[feature].dropna(), bins=20, color='pink', alpha=0.6)
    plt.title(feature.replace("_", " ").title())
    plt.tight_layout()
plt.show()
```

```python
correlation_matrix = red_wine_data.corr()
plt.figure(figsize=(10, 8))
plt.title('CORRELATION MATRIX')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.7)
plt.show()
```
